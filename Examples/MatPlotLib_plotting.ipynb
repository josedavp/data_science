```python
#### line graphs using GDP data


years = list(range(1950, 2011, 10))
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create a line chart, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# add a title
plt.title("Nominal GDP")

# add a label to the y-axis
plt.ylabel("Billions of $")

# add a label to the x-axis
plt.xlabel("Year")

plt.show()
```

```python
##

years = list(range(1950, 2011, 10))
gdp1 = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
gdp2 = [226.0, 362.0, 928.0, 1992.0, 4931.0, 7488.0, 12147.0]
gdp3 = [1206.0, 1057.0, 1081.0, 2940.0, 8813.0, 13502.0, 19218.0]
gdp4 = [500, 501, 502, 504, 508, 514, 524]
# create a line chart, years on x-axis, gdp on y-axis
# use format string to specify marker and line styles
plt.plot(years, gdp1, 'o-', years, gdp2, '*:', years, gdp3, 'd-.')
# add a title
plt.title("Nominal GDP")
# add a label to the y-axis
plt.ylabel("Billions of $")
# add a label to the x-axis
plt.xlabel("Year")
# add legend
plt.legend(['countryA', 'countryB', 'countryC'])
plt.show()

```

```python
##

gdp4 = [500, 501, 502, 504, 508, 514, 524]

plt.plot(years, gdp4, 'c<--')
# add a title
plt.title("Nominal GDP")
# add a label to the y-axis
plt.ylabel("Billions of $")
# add a label to the x-axis
plt.xlabel("Year")
# add legend
plt.legend(['countryD'])
plt.show()

plt.plot(years, gdp1, 'o-', years, gdp2, '*:', years, gdp3, 'd-.')
plt.title("Nominal GDP")
plt.ylabel("Billions of $")
plt.xlabel("Year")
plt.legend(['countryA', 'countryB', 'countryC'])

plt.show()

```

## How to interpret the graphs above?

1. How does the growth rate of Country D compare with other contries?

2. How does the growth rate of the four contries during the period from 1950 to 1970?


```python
##

plt.plot(years, gdp1, 'o-', years, gdp2, '*:', years, gdp3, 'd-.')
plt.plot(years, gdp4, 'c<--')
# add a title
plt.title("Nominal GDP")
# add a label to the y-axis
plt.ylabel("Billions of $")
# add a label to the x-axis
plt.xlabel("Year")
# add legend
plt.legend(['countryA', 'countryB', 'countryC', 'countryD'])
plt.show()

```

```python
##

plt.plot(years, gdp4, 'c--<'); 
plt.ylim([0, 1000])
plt.annotate('plt.ylim([0, 1000])', xy=(1950, 600), xytext=(1960, 650), fontsize=20, arrowprops=dict(facecolor='black', shrink=0.05))
plt.title("Nominal GDP")
plt.ylabel("Billions of $")
plt.xlabel("Year")
plt.legend(['countryD'])
plt.show()

```

```python
##

plt.semilogy(years, gdp1, 'bo-', 
             years, gdp2, 'r*:', 
             years, gdp3, 'gd-.', 
             years, gdp4, 'c--<')
# add a title
plt.title("Nominal GDP")
# add a label to the y-axis
plt.ylabel("Billions of $")
# add a label to the x-axis
plt.xlabel("Year")
# add legend
plt.legend(['countryA', 'countryB', 'countryC', 'countryD'])
plt.show()

```

```python
## line graph on final exam grades
random.seed(5)
grades = random.randint(50, 100, 20)
plt.plot(range(1,21), grades, '-*')
plt.xticks(range(0,25,5));
plt.xlabel('Student')
plt.ylabel('Grade')
plt.title('Final exam grade')
plt.show()

```

```python
##

plt.plot(range(1,21), sorted(grades), '-*')
plt.xticks(range(0,25,5));
plt.xlabel('Student')
plt.ylabel('Grade')
plt.title('Final exam grade after sorting')
plt.show()

```

```python
## line graph on running time


seconds = (random.randn(200)+10)*20
plt.plot(range(1,201), seconds, '-')
plt.xticks(range(0,205,20));
plt.xlabel('Dataset')
plt.ylabel('Running time (seconds)')
plt.title('Running time of program on random data')
plt.show()

```

```python
##

plt.plot(range(1,201), sorted(seconds), '-')
plt.xticks(range(0,205,20));
plt.xlabel('Dataset')
plt.ylabel('Running time (seconds)')
plt.title('Sorted running time')
plt.show()

```

```python
#### bar charts

movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

xs = range(len(movies)) # xs is range(5)
# plot bars with left x-coordinates [xs], heights [num_oscars]
plt.bar(xs, num_oscars)
# label x-axis with movie names at bar centers
plt.xticks(xs, movies)
# alternatively, use the following to replace the two lines above
#plt.bar(xs, num_oscars, tick_label=movies)

plt.ylabel("# of Academy Awards")
plt.title("My Favorite Movies")
plt.show()

```

```python

##

plt.bar(range(len(years)), gdp4, tick_label=years);
plt.xlabel('Year')
plt.ylabel('Nominal GDP')
plt.title("Nominal GDP")
plt.show()

```

```python
##

plt.barh(xs, num_oscars, tick_label=movies)
plt.xlabel("# of Academy Awards")
plt.title("My Favorite Movies")
plt.show()
```

```python
##### scatter plots

friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

# label each point
for i in range(len(friends)):
    plt.annotate(labels[i], xy=(friends[i], minutes[i]), xytext=(5, 0), textcoords='offset points')

plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()

```

```python

###### subplots

fig=plt.figure() # create a new figure

fig.add_subplot(2,2,1);
plt.plot(years, gdp1);
plt.ylabel("Billions of $")

fig.add_subplot(2,2,2);
plt.bar(xs, num_oscars, tick_label=movies)
plt.ylabel("# of Academy Awards")


fig.add_subplot(2,2,3);
plt.scatter(friends, minutes)

# label each point
for i in range(len(friends)):
    plt.annotate(labels[i], xy=(friends[i], minutes[i]), xytext=(5, 0), textcoords='offset points')

plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()
```
